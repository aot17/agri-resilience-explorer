```python
import ee

ee.Authenticate()
ee.Initialize()

print("Earth Engine initialized successfully")
```

```python
ee.Initialize()

point = ee.Geometry.Point([37.65, -1.5])  # Kenya

s2_small = (
    ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED")
    .filterBounds(point)
    .filterDate("2024-01-01", "2024-01-31")
)

print(s2_small.size().getInfo())
```

```python
image = s2_small.sort("CLOUDY_PIXEL_PERCENTAGE").first()

print(image.get("CLOUDY_PIXEL_PERCENTAGE").getInfo())
```

```python
import geemap

Map = geemap.Map(center=[-1.5, 37.65], zoom=10)

vis = {
    "bands": ["B4", "B3", "B2"],
    "min": 0,
    "max": 3000
}

Map.addLayer(image, vis, "Sentinel-2 RGB")
Map
```
